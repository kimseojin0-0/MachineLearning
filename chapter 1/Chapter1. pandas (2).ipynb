### DataFrame의 칼럼 데이터 세트 생성과 수정

```python
import pandas as pd

titanic_df = pd.read_csv('./titanic_train.csv')
titanic_df
```

```python
#새로운 칼럼 추가
titanic_df['Age_0'] = 0
titanic_df.head(3)
```

```python
#기존 칼럼 series의 데이터로 새로운 칼럼의 series만들기
titanic_df['Age_by_10'] = titanic_df['Age']*10
titanic_df['Family_No'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df.head(3)
```

```python
#기존 칼럼값 업데이트하기
titanic_df['Age_by_10'] = titanic_df['Age_by_10'] + 100
titanic_df.head(3)
```

### DataFrame 데이터 삭제

- drop(labels=원하는 칼럼명, axis=0 또는 1, ... , inplace=False)


- axis=0으로 설정하고 로우 레벨로 삭제하는 경우, 이상치 데이터 삭제할 때 주로 사용

```python
titanic_drop_df = titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
```

```python
titanic_df.head(3)
#Age_0 칼럼이 삭제되지 않음
```

- 삭제되지 않은 이유: inplace를 따로 지정하지 않으면 디폴트값이 False이기 때문에


- inplace=False는 자기 자신의 df 데이터는 삭제하지 않고, 삭제된 결과만 반환함

```python
titanic_df.head()
```

```python
drop_result= titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
print('inplace=True 로 drop한 후 반환된 값:', drop_result)
titanic_df.head(3)
```

```python
#axis=0으로 설정헤 맨 앞 3개 데이터 삭제하기
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 15)
print('#### before aixs 0 drop ####')
print(titanic_df.head(3))

titanic_df.drop([0, 1, 2], axis=0, inplace=True)

print('#### after axis 0 drop ####')
print(titanic_df.head(3))
```
